```python
cotton_clothes = fashion_df[fashion_df['name'].str.contains('Cotton', case=False)]
woolen_clothes = fashion_df[fashion_df['name'].str.contains('Woolen', case=False)]
rainy_clothes = fashion_df[fashion_df['name'].str.contains('Raincoat', case=False)]
regular_clothes = fashion_df

fashion_df.head()
```

```python
import pandas as pd

# Assuming you have pre-filtered your clothing dataset like this
# cotton_clothes, woolen_clothes, rainy_clothes, regular_clothes have been filtered earlier

def recommend_clothing(temp, season, weather_type):
    if temp >= 30 and weather_type.lower() == "sunny":
        # Recommend cotton clothes for hot sunny weather
        return cotton_clothes.sample(n=5).to_dict(orient='records')
    
    elif temp >= 30 and weather_type.lower() in ["cloudy", "partly cloudy"]:
        # Recommend lighter cotton or casual clothes for hot cloudy weather
        return cotton_clothes.sample(n=5).to_dict(orient='records')
    
    elif temp <= 15 and season.lower() == "winter":
        # Recommend woolen clothes for cold winter weather
        return woolen_clothes.sample(n=5).to_dict(orient='records')
    
    elif temp <= 15 and weather_type.lower() == "rainy":
        # Recommend warm clothes for cold rainy weather
        return woolen_clothes.sample(n=5).to_dict(orient='records')
    
    elif 15 < temp < 30 and weather_type.lower() == "rainy":
        # Recommend clothes for rainy weather (not too cold, not too hot)
        return rainy_clothes.sample(n=5).to_dict(orient='records')
    
    elif temp >= 20 and season.lower() == "spring":
        # Recommend light spring clothes
        return cotton_clothes.sample(n=5).to_dict(orient='records')

    elif temp >= 20 and season.lower() == "summer":
        # Recommend summer clothes for hot weather
        return cotton_clothes.sample(n=5).to_dict(orient='records')

    elif 10 < temp < 20 and weather_type.lower() == "windy":
        # Recommend clothes for windy conditions
        return regular_clothes.sample(n=5).to_dict(orient='records')

    else:
        # Default: Recommend general casual clothes if no specific condition matches
        return regular_clothes.sample(n=5).to_dict(orient='records')
```

```python
weather_df['name'] = weather_df.apply(
    lambda row: recommend_clothing(row['Temperature'], row['Weather Type'], row['Season']), axis=1
)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.3,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
one_hot_encoder = OneHotEncoder(handle_unknown='ignore')
categorical_columns = ['Season', 'Weather Type', 'Location', 'Cloud Cover']

# Define the preprocessor for categorical columns
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', one_hot_encoder, categorical_columns)  # Use your categorical column names here
    ],
    remainder='passthrough'  # Keep the remaining numerical columns as they are
)
```

```python
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', KNeighborsClassifier(n_neighbors=5))  # KNN with 5 neighbors
])
```

```python
model_pipeline.fit(X_train, Y_train)
```

```python
accuracy = model_pipeline.score(X_test, Y_test)
print(f"Model Accuracy: {accuracy:.2f}")
```
